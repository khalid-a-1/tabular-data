```python
"""The code below reads two .csv files into data frames using pandas."""
import pandas as pd
#we are importing the pandas library to use in out code 

weather = pd.read_csv("weather.csv")
flights = pd.read_csv("flights.csv")
#we open both the weather and flights csv files and make them into dataframes
#we then assign the varibles weather and flights to respected dataframes 
```

## *Weather Tidy Data:*

```python
"""The code below is to tidy up the weather.csv data and making that tidied data into a new .csv file."""

weather.T
#we transpose the dataframe 

melted = weather.melt(id_vars=['year','month','measure'], var_name='Day', value_name='V')
#we melt the dataframe which converts rows into columns in a specific way

melted.dropna(axis=0,inplace=True)
#we use the .dropna() to operation to remove all the entries that are null

x = lambda y: y != 'T' 
melted['N'] = melted['V'].apply(x)
melted =melted.loc[melted['N']]
melted.drop(['N'], axis=1, inplace=True)
#we use a lambda expression to find a string entry and we make a new column and used in .apply()
#on a column that we created in .melt() and we further drop the extra column we created to 
#store values 

xv = lambda y: y != 'X' 
melted['X'] = melted['Day'].apply(xv)
melted =melted.loc[melted['X']]
melted.drop(['X'], axis=1, inplace=True)
#we do the same process as the one above but on the 'Day' column 

c = lambda b: int(b[1:])
melted['Day'] = melted['Day'].apply(c)
#we use a lambda expression to convert a string into integer and only keep the number part 
#in the 'Day' column

melted = melted.pivot_table(index=['year','month', 'Day'], columns= 'measure', values= 'V', aggfunc= 'first')
#we pivot our table to make the 'measure' into columns while indexing 'year','month' and'Day'

melted = melted.sort_values(by= ['year','month', 'Day'])
#we sort the values by the date using .sort_values() 

melted["Events"].fillna("No data", inplace = True)
#we fill in the void entries in the Events column with "No Event" using .fillna()

melted.to_csv('weatherTidy.csv', sep='\t', encoding='utf-8')
#we make our new tidy dataframe into a csv file

melted.head()
#visible representation of the new tidy data
```

In the tidying of the weather data, I start off by using transpose to make it more accessible to use *.melt()* on our data frame. When I use melt on the data frame, I declare certain columns as my index and then make my column of 'measure' rows into columns and assign them their specific values. I further use *.pivot_table()* to get the values into specific columns while they are in specific rows aligning with their indexes. I then use expressions to remove the values from my data that are 'T' as those values are of no use to us. To make my Day column tidier, I remove the substring 'X' and convert the remaining string into an integer to make more sense of it as a day. I sort the values in my data frame using *.sort_values()*. I sort on the bases of year, month and day columns. I further removed the data that had unrecorded values by using *.dropna()* on the data frame. I also filled the unrecorded values in the Events column as 'No data' because we further see that the events column can still be useful. I exported my tidied data frame into a *.csv* file so it can be used in tableau.  

## *Flights Tidy Data:*

```python
"""The code below is to tidy up the flights.csv data and making that tidied data into a new .csv file."""

flights.drop(['time_hour','hour','minute'], axis=1, inplace=True)
#we use .drop() to drop particular columns

flights.set_index(['year', 'month', 'day','origin','dest','dep_time'], inplace=True)
#we set out date, origin and destination as index 

flights.sort_values(by = ['dep_delay'], inplace=True)
#sorting values using departure delay

flights.to_csv('flightsTidy.csv', sep='\t', encoding='utf-8')
#we make our new tidy dataframe into a csv file

flights.head()
#visible representation of the new tidy data
```

In tidying of the flights data, I omit the 'time_hour', 'hour' and 'minute' columns since they are repeating the same information that we have in our other columns for example the contents of 'time_hour' are already stored in our year, month, day, and sched_dep columns. We set our index as 'year', 'month', 'day','origin','dest' and 'dep_time' columns to meet the tidy constraints. We sorted the data using our 'dep_delay' column because it more useful sorted in exploring the data. We export my tidied data into a *.csv* file so it can be used in tableau.
